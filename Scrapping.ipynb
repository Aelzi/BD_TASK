```python
# Buat header untuk file CSV

kepala=['Symbol','MarketCap','Price','Circulating Supply','Volume(24h)','%1H','%24H','%7D', 'Date']

with open('FIXINIMAH.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(kepala)

# writer = csv.writer(open('asu.csv', 'w', newline=''))
# writer.writerow(kepala)

# Tentukan tanggal awal dan akhir
    start_date = datetime(2018, 1, 1)
    end_date = datetime(2023, 9, 1)

    # URL dasar
    base_url = 'https://coinmarketcap.com/historical/'

    # Daftar untuk menyimpan URL yang dihasilkan
    url_list = []

    # Loop melalui tanggal-tanggal dari awal hingga akhir
    current_date = start_date
    while current_date <= end_date:
        # Format tanggal dalam format YYYYMMDD
        date_str = current_date.strftime('%Y%m%d')
        
        # Buat URL dengan tanggal yang diformat
        url = base_url + date_str + '/'
        
        # Tambahkan URL ke daftar URL
        url_list.append(url)
        
        # Tambahkan satu hari ke tanggal saat ini
        current_date += timedelta(days=1)


    # Buat daftar untuk menyimpan data
    datas = []

    # Loop melalui daftar URL
    for url in url_list:
        print('scraping page:', url)
        req = requests.get(url)
        soup = BeautifulSoup(req.text, 'html.parser')
        items = soup.findAll('tr', 'cmc-table-row')
        

        for it in items:

            nama = it.find('td', 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--left cmc-table__cell--sort-by__symbol').text
            if nama in ['BTC', 'ETH', 'BNB', 'ADA', 'USDT', 'SOL', 'XRP', 'DOT', 'LTC']:

                marketCap = it.find('td', 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__market-cap').text.replace('$', '').replace(',', '')
                try: marketCap = float(marketCap)
                except: marketCap=marketCap
                        
                price = it.find('td', 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__price').text.replace('$', '').replace(',', '')
                try:price = float(price)
                except: price=price
                
                circulatingSupply = it.find('td', 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__circulating-supply').text.replace(',', '').split(' ')
                try:circulatingSupply = int(circulatingSupply[0])
                except: circulatingSupply=circulatingSupply
                
                volume24h = it.find('td', 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__volume-24-h').text.replace('$', '').replace(',', '')
                try:volume24h = float(volume24h)
                except:  volume24h=volume24h

                percentChange1H = it.find('td', 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__percent-change-1-h').text.replace('%', '').replace(',', '.').replace('<', '0')
                try: percentChange1H = float(percentChange1H)
                except:  percentChange1H=percentChange1H
                
                percentChange24H = it.find('td', 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__percent-change-24-h').text.replace('%', '').replace(',', '.').replace('<', '0')
                try: percentChange24H = float(percentChange24H)
                except: percentChange24H=percentChange24H
                
                percentChange7D = it.find('td', 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__percent-change-7-d').text.replace('%', '').replace(',', '.').replace('<', '0')
                try: percentChange7D = float(percentChange7D)
                except: percentChange7D=percentChange7D
                
                # Menambahkan data ke daftar datas
                datas.append([nama, marketCap, price, circulatingSupply, volume24h, percentChange1H, percentChange24H, percentChange7D, url[-9:-1]])
            else:
                break



# Menyimpan data ke dalam file CSV

    for d in datas:
        writer.writerow(d)

print('Selesai')



# for d in datas:
#     writer.writerow(d)
    
# csv_file.close()
    
# df = pd.read_csv('as.csv')
```
